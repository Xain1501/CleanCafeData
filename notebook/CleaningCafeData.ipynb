CLEANING DIRTY CAFE DATA

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df =pd.read_csv('dirty_cafe_sales.csv')
```

```python
df.head(20)
```

```python
df.isna().sum()
```

```python
df.nunique()
```

```python
df['Location']= df['Location'].astype(str).str.lower().str.strip()
df['Location']=df['Location'].replace(['unknown', 'error', 'nan', 'n/a', 'none'], np.nan)
df['Location'].isna().sum()
df['Location'] = df['Location'].fillna('Not Recorded')
```

```python
df.head(20)
```

```python
df['Total Spent']= df['Total Spent'].astype(str).str.lower().str.strip()
df['Total Spent']=df['Total Spent'].replace(['unknown', 'error', 'nan', 'n/a', 'none'], np.nan)
df['Total Spent'].isna().sum()
```
